```python
#Problem 1
class bankaccount:
  def __init__(self,acno,achn,bal,withdraw,depositamt):
    self.acno=acno
    self.achn=achn
    self.bal=bal
    self.depositamt=depositamt
    self.withdraw=withdraw
  def accountnumber(self):
    print("This is your account number:",self.acno)
  def accountholdername(self):
    print("This is your name:",self.achn)
  def balance(self):
    print("Amt in your account=",self.bal)
  def deposit(self):
    print("The amount has been deposited, balance now:",self.bal+self.depositamt)
  def withdraw(self):
    print("The amount has been withdrawn, balance now:",self.bal-self.withdraw)
```

```python
#Problem 6
class shape_calculation():
  def __init__(self,length,breadth,height):
    self.len=length
    self.bth=breadth
    self.ht=height
  def area(self):
    print("area of shape:",self.len*self.bth)
  def volume(self):
    print("volume of shape:", self.len*self.bth*self.ht)
```
